# Deskripsi Kasus

Data yang digunakan merupakan kumpulan data yang mencakup 3 skor atau nilai tes siswa di sekolah umum serta berbagai faktor pribadi dan sosial-ekonomi yang mungkin memiliki efek interaksi pada mereka. Data tersebut digunakan untuk mengetahui faktor apa yang mempengaruhi nilai tes siswa di sekolah, yakni sebagai berikut:

1. Gender: jenis kelamin siswa (laki-laki/perempuan)

2. Ethnic Group: kelompok etnis siswa (kelompok A - E)

3. Parent Education: latar belakang pendidikan orang tua siswa (dari some_highschool sampai master's degree)

4. Lunch type: jenis makan siang siswa (standar/kurang)

5. Test preparation: kursus persiapan tes yang diikuti siswa (selesai/ tidak sama sekali)

6. Parent marital status: status perkawinan orang tua (menikah/lajang/janda/bercerai)

7. Practice sport: seberapa sering siswa latihan olahraga (tidak pernah/kadang/teratur)

8. Is first child: apakah merupakan anak pertama dalam keluarga (ya/tidak)

9. Nr siblings: jumlah saudara kandung yang dimiliki siswa (0-7)

10. Transport means: sarana transportasi ke sekolah (bus sekolah/pribadi)

11. Weekly study hours: jam belajar mandiri mingguan siswa (kurang dri 5 jam/ 5-10/ lebih dari 10 jam)

12. Math score: nilai tes matematika (0-100)

13. Reading score: nilai tes membaca (0-100)

14. Writing score: nilai tes menulis (0-100)

```python
import pandas as pd

df = pd.read_csv(r'D:\mata kuliah\analisis data eksploratif\Expanded_data_with_more_features.csv', sep= ",")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#Visualisasi Gender
plt.figure(figsize=(4, 3))
sns.countplot(x='Gender', data=df, palette='Set2')
plt.title('Distribusi Gender')
plt.show()

# Visualisasi Ethnic Group
plt.figure(figsize=(6, 4))
sns.countplot(x='EthnicGroup', data=df, palette='Set1')
plt.title('Distribusi Kelompok Etnis Siswa')
plt.show()

# Visualisasi Parent Education
plt.figure(figsize=(8, 5))
sns.countplot(y='ParentEduc', data=df, palette='Set3', order=df['ParentEduc'].value_counts().index)
plt.title('Tingkat Pendidikan Orang Tua Siswa')
plt.show()

# Visualisasi Lunch Type
plt.figure(figsize=(4, 3))
sns.countplot(x='LunchType', data=df, palette='Set2')
plt.title('Distribusi Jenis Makan Siang Siswa')
plt.show()

# Visualisasi Test Preparation
plt.figure(figsize=(4, 3))
sns.countplot(x='TestPrep', data=df, palette='Set1')
plt.title('Distribusi Kursus Persiapan Tes')
plt.show()

# Visualisasi Parent Marital Status
plt.figure(figsize=(6, 4))
sns.countplot(x='ParentMaritalStatus', data=df, palette='Set2', order=df['ParentMaritalStatus'].value_counts().index)
plt.title('Distribusi Status Perkawinan Orang Tua')
plt.xticks(rotation=45)
plt.show()

# Visualisasi Practice Sport
plt.figure(figsize=(4, 3))
sns.countplot(x='PracticeSport', data=df, palette='Set1', order=df['PracticeSport'].value_counts().index)
plt.title('Distribusi Frekuensi Latihan Olahraga Siswa')
plt.show()

# Visualisasi Is First Child
plt.figure(figsize=(4, 3))
sns.countplot(x='IsFirstChild', data=df, palette='Set2')
plt.title('Distribusi Anak Pertama atau Bukan')
plt.show()

# Visualisasi Nr Siblings
plt.figure(figsize=(6, 4))
sns.histplot(df['NrSiblings'], bins=range(0, 8), kde=False, color='skyblue')
plt.title('Distribusi Jumlah Saudara Kandung')
plt.xlabel('Jumlah Saudara Kandung')
plt.ylabel('Jumlah Siswa')
plt.show()

# Visualisasi Transport Means
plt.figure(figsize=(6, 4))
sns.countplot(x='TransportMeans', data=df, palette='Set1')
plt.title('Distribusi Sarana Transportasi ke Sekolah')
plt.show()

# Visualisasi Weekly Study Hours
plt.figure(figsize=(6, 4))
sns.countplot(x='WklyStudyHours', data=df, palette='Set2')
plt.title('Distribusi Jam Belajar Mandiri Mingguan Siswa')
plt.xlabel('Jam Belajar Mandiri')
plt.ylabel('Jumlah Siswa')
plt.xticks(rotation=45)  # Rotasi label agar lebih mudah dibaca
plt.show()

plt.figure(figsize=(8, 6))
sns.histplot(df['MathScore'], bins=10, kde=True, color='blue')
plt.title('Distribusi Nilai Ujian Matematika')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.xlim(0, 100)
plt.show()

plt.figure(figsize=(8, 6))
sns.histplot(df['ReadingScore'], bins=10, kde=True, color='blue')
plt.title('Distribusi Nilai Ujian Membaca')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.xlim(0, 100)
plt.show()

plt.figure(figsize=(8, 6))
sns.histplot(df['WritingScore'], bins=10, kde=True, color='blue')
plt.title('Distribusi Nilai Ujian Menulis')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.xlim(0, 100)
plt.show()
```
